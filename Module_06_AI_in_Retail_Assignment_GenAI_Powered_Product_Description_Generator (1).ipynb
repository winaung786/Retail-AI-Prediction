```python
from flask import Flask, request, render_template
from transformers import pipeline

app = Flask(__name__)

# Load the GPT-2 model for text generation
generator = pipeline('text-generation', model='gpt2')

@app.route('/', methods=['GET', 'POST'])
def index():
    description = ""
    if request.method == 'POST':
        product_name = request.form.get('product_name')
        keywords = request.form.get('keywords')
        # Generate product description using GPT-2
        prompt = f"Product Name: {product_name}\nKeywords: {keywords}\nDescription:"
        description = generator(prompt, max_length=100, num_return_sequences=1)[0]['generated_text']
    return render_template('index.html', description=description)

if __name__ == '__main__':
    app.run(debug=True)
```
